description: This dataset recorded 3000 music therapy sessions from about 500 patients, and involved the parameters of the music used, and the change of patient’s mental and physical health.

---

# 3. Initial dataset review

Number of rows: 300  
Number of columns: 16  

---

## Columns name and explanations:

- Patient id: record the id of each patient, nominal value  
- Session id: record the id of each therapy session, nominal value  
- Diagnostics: record the type of mental problem of each patient and the degree of severity, nominal value  
- Music type: record the type of music used in the therapy, nominal value  
- Duration_min: record the duration of therapy in minutes, ratio value  
- HR_temporal: heart rate measured during therapy (beat per minute), ratio value  
- HRV temporal: heart rate variability indicating nervous system response, interval value  
- Respiration_temporal: breathing rate recorded during the session, ratio value  
- GSR_temporal: Galvanic Skin Response reflecting emotional arousal, interval value  
- Tempo_spectral: tempo feature of the music in beats per minute, ratio value  

---

## Missing value:

Repetition: This index can be used to measure the degree of repetition of the music. If the music is more repetitive, it will probably have a better effect on calming the patients down. (Interval value)  
Volume: this dataset doesn’t include the volume of the music, which may affect the effectiveness of music therapy. (ratio value)
